```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('wdi_data.csv')
df.head()
```

```python
df = df[df["indicator"] =="SP.POP.TOTL"]
df = df[df["Country Code"] =="USA"]
df.head(20)
```

```python
fig = go.Figure()
trace_0 = go.Bar(
    x=df["Year"],
    y=df["value"]
)
fig.add_trace(trace_0)

fig.show()
```

```python
fig.update_layout(
    title ="USA Population Growth",
    xaxis ={"title":"year"},
    yaxis ={"title" :"population"}
)
fig.show()
```

```python
fig = go.Figure()
trace_0 = go.Line(
    x=df["Year"],
    y=df["value"],
    mode="markers+lines"
)
fig.add_trace(trace_0)

fig.show()
fig.update_layout(
    title ="USA Population Growth",
    xaxis ={"title":"year"},
    yaxis ={"title" :"population"}
)
fig.show()
```
