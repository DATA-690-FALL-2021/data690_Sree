# 6.Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function

```python
df.dtypes
```

```python
def cleanse_year(i):
    a = i.replace("*","")
    return a
df["year"] = df["Census Year"].apply(cleanse_year)
df["year"] = pd.to_numeric(df["year"])
df
```

```python
def cleanse_pop(i):
    a = i.replace(",","")
    return a
df["pop"] = df["Total Population"].apply(cleanse_pop)
df["pop"] = pd.to_numeric(df["pop"])
df
```

```python
def cleanse_cost(i):
    return (i.replace(",","").replace("$","").replace(" Billion","000000000"))
df["cost"] = df["Census Cost"].apply(cleanse_cost)
df["cost"] = pd.to_numeric(df["cost"])
df
```

```python
def cleanse_avg(i):
    return (i.replace(",","").replace("$","").replace(" cents",""))
df["avg_cost"] = df["Average Cost Per Person"].apply(cleanse_avg)
df["avg_cost"] = pd.to_numeric(df["avg_cost"])
df
```
