#10.7 Find and display Johns Hopskins


```python
findRec4 = df1.loc[df1['INSTNM'] == 'Johns Hopkins University']
findRec4
```

#10.8 Calculate the min, max, and average in-state tution of the above four institutions.

```python
final_df = pd.concat([findRec1, findRec2, findRec3, findRec4], axis=0)
final_df
```

```python
min = final_df['TUITIONFEE_IN'].min()
print("Minimum In-state Tuition: " ,min)

max = final_df['TUITIONFEE_IN'].max()
print("Maximum In-state Tution: " ,max)

avg = final_df['TUITIONFEE_IN'].mean()
print("Avgerage In-state Tuition:", avg)
```
